# Downloading data

```python
from fastai.text.all import *
path = untar_data(URLs.WIKITEXT_TINY) #downloading WIKI_PEDIA dataset from fastai
```

# Data preprocessing

```python
import pandas as pd

train_df = pd.read_csv(path/'train.csv', header=None)#load the data without header
test_df = pd.read_csv(path/'test.csv', header=None)
train_df.columns = ['text'] #add text for header
test_df.columns = ['text']

print(train_df.head())  
```

```python
from fastai.text.all import *

# Create our DataLoader
dls = TextDataLoaders.from_df(
    train_df,
    text_col='text',
    is_lm=True,       # Indicates it's for language modeling
    bs=16,            # Batch size
    seq_len=64        # Sequence length of each training example
)
```

# Run and validation

```python
learn = language_model_learner(dls, AWD_LSTM, drop_mult=0.5,wd=0.1, metrics=[accuracy, Perplexity()]).to_fp16()
learn.fit_one_cycle(2, lr_max=1e-3, moms=(0.8, 0.7,0.8), cbs=GradientClip(0.1))
```

```python
### this is an excelent  result base on 22 perplexity and dou to hardness of next word prediction problems now is time to use it and see how it works
```

```python
Text='Lady Rosebery '
N_sentences=20
N_words=50
preds=[]
for i in range(N_sentences):
    preds.append(learn.predict(Text,N_words,temperature=0.75))
for j in preds:
    print(j)
```
